```python
## To create a dataframe out of the dictionary
bson_df= pd.DataFrame(data=boston.data,columns=boston.feature_names)
bson_df
```

```python
#Created a new colun with name MEDV and passed the target values from the dictionary
bson_df['MEDV'] = boston.target
bson_df.head()
```

```python
## Out of 13 columns --> today we will work with
# single input and medv as output
```

```python
## We are using RM columns as input(SLR--> we have only one input)
## Output--> MEDV column
```

```python
# Preparing the Input and Output columns
X = bson_df[['RM']]
y=bson_df[['MEDV']]
```

```python
bson_df['MEDV']
```

```python
#plt.style.available
with plt.style.context("grayscale"):
    plt.plot(X,y,"bo")
    plt.xlabel("RM I/P")
    plt.ylabel("MEDV O/P")
    plt.title("Scatter Plot")
```

```python
## Usual sklearn: import ,instantiate,fit
from sklearn.linear_model import LinearRegression
lm =LinearRegression()
lm.fit(X,y)
```

```python
## Print intercept and coefficient
print(lm.intercept_)
print(lm.coef_)
```

```python
# y=mx+c
#m value--> coefficient(9.10210898)
#c value--> intercept(-34.67062078)
(9.10210898 * 6.575) + (-34.67062078)
```

### Interpeting the model

A unit increase in X(RM) is associated with a 9.10210898 unit increae in MEDV

```python
## Pipeline--> First we create a model on the data
## Second, we need to compare the model(prediction values with actual values)
```

```python
## With help of predict ,we are going to find our predicted values 
y_pred=lm.predict(X)
y_pred
```

```python
## Compare the actual with the predicted values
y==y_pred
```

```python
## r2_value--> to check the model performance
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```

```python
#Observe the regression line
with sns.axes_style('dark'):
    sns.regplot(X,y)
```

```python
## scatterplot(Regression)
# First we are making a scatter plot of the original data(X and Y)
# We need to plot a line for the predicted values(X,y_pred)
plt.scatter(X,y,color="red")
plt.plot(X,y_pred,color="black")
plt.xlabel("RM I/P")
plt.ylabel("MEDV O/P")
```
